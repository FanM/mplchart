# A quick tour of builtin indicators

```python
import yfinance as yf

from mplchart.chart import Chart

from mplchart.primitives import Candlesticks, Volume, SameAxes
from mplchart.indicators import ROC, SMA, EMA, RSI, MACD, SLOPE, BBANDS

```

```python
ticker = 'AAPL'
prices = yf.Ticker(ticker).history('5y')
```

```python
max_bars = 250

indicators = [
    Candlesticks(), SMA(50), SMA(200), Volume(),
    RSI(),
    MACD(),
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
max_bars = 250
indicators = [ Candlesticks(), SLOPE(20), SameAxes(), SLOPE(50) ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
max_bars = 250
indicators = [ Candlesticks(), BBANDS() ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
