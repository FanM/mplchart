```python
from mplchart.chart import Chart
from mplchart.helper import get_prices

from mplchart.primitives import Candlesticks, Volume, Peaks
from mplchart.indicators import EMA, RSI, MACD

```

```python
ticker = 'NVDA'
freq = 'daily'
prices = get_prices(ticker, freq=freq)
prices.tail()
```

```python
%%time

max_bars = 260
indicators = [Candlesticks() , EMA(12), Peaks(3), Volume(),
              RSI(),
              MACD(), Peaks(3, item='macd'),
              ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
