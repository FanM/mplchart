# Multiple Asset Plots (tentative)

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import OHLC, Candlesticks, Volume, Price
from mplchart.indicators import SMA, RSI, MACD
```

## Fetching the prices

```python
ticker = 'AAPL'
aapl = yf.Ticker(ticker).history(period='5y')
aapl
```

```python
ticker = 'NVDA'
nvda = yf.Ticker(ticker).history(period='5y')
nvda
```

## Creating the Chart

```python
# Multiple plots without rebase

style = 'tableau'
max_bars = 250
title = "AAPL vs NVDA"

indicators = [
    Candlesticks(), Volume(),
    RSI(),
    MACD(),
]

chart = Chart(title=title, max_bars=max_bars, style=style)
chart.plot(aapl, indicators)
chart.plot(nvda, [ Price() ] , target='main')
```

```python
# Multiple plots with rebase

chart = Chart(title=title, max_bars=max_bars, style=style)
chart.plot(aapl, indicators)
chart.plot(nvda, [ Price() ] , target='main', rebase=True)
```
