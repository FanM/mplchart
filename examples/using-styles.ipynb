# Using mplchart with matplotlib styles

```python
import yfinance as yf
import matplotlib.pyplot as plt

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume
from mplchart.indicators import SMA, RSI, MACD
```

```python
IGNORE = (("_", "seaborn"))
styles = [ s for s in plt.style.available
          if not s.startswith(IGNORE)]

styles
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    SMA(50),
    SMA(200),
    Volume(),
    RSI(),
    MACD(),
]

chart = Chart(title="default", max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
IGNORE = (("_", "seaborn"))
styles = [ s for s in plt.style.available
          if not s.startswith(IGNORE)]


for style in styles:
    with plt.style.context(style):
        chart = Chart(title=style, max_bars=max_bars)
        chart.plot(prices, indicators)
        chart.show()
```
