# Chart Gallery

```python
import yfinance as yf

from mplchart.chart import Chart

from mplchart.indicators import SMA, EMA, ROC, MACD, PPO, SLOPE

from mplchart.primitives import  Candlesticks, OHLC, Price, Volume, SameAxes, LinePlot
```

## Fetching the prices

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

## Candlestick plot

```python
max_bars = 250
indicators = [Candlesticks()]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Open High Low Close bars

```python
max_bars = 250
indicators = [OHLC()]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Price plot

```python
max_bars = 250
indicators = [Price()]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Candlesticks and Volume

```python
max_bars = 250
indicators = [Candlesticks(), Volume()]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Plotting in the same pane with SameAxes primitive

```python
max_bars = 250

indicators = [
    Candlesticks(),
    ROC(20),
    ROC(50) | SameAxes(),
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```

## Customizing style with LinePlot

```python
max_bars = 250

indicators = [
    Candlesticks(),
    SMA(20) | LinePlot(style="dashed", color="red", alpha=0.5, width=3)
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```
