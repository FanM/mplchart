# Chart Primitives

```python
import yfinance as yf

from mplchart.chart import Chart

from mplchart.primitives import (
    Candlesticks,
    OHLC,
    Price,
    Volume,
    Peaks,
)
from mplchart.indicators import SMA, EMA, ROC, MACD, PPO, SLOPE
```

## Fetching the prices

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

## Candlestick plot

```python
max_bars = 250
indicators = [Candlesticks()]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Open High Low Close bars

```python
max_bars = 250
indicators = [OHLC()]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Price plot

```python
max_bars = 250
indicators = [Price()]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Volume plot

```python
max_bars = 250
indicators = [Volume(50)]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Candlesticks and Volume

```python
max_bars = 250
indicators = [Candlesticks(), Volume()]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```
