# Candlesticks Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Candlesticks (default)

max_bars = 250

indicators = [ Candlesticks() ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
# Candlesticks (using bars) 

max_bars = 250

indicators = [ Candlesticks(use_bars=True) ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
# Candlestik colors

max_bars = 250

indicators = [ Candlesticks(colorup="green", colordn="red") ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
# Candlesticsk colors (using bars)

max_bars = 250

indicators = [ Candlesticks(colorup="green", colordn="red", use_bars=True) ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```
