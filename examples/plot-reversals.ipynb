```python
import yfinance as yf
from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume, Reversal
from mplchart.indicators import RSI
from auto_chart_patterns.reversal_patterns import ReversalPatternProperties
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="1y")
```

```python
# Configure pattern scanning
scan_props = ReversalPatternProperties(
    flat_ratio=0.2,
)


indicators = [
    Candlesticks(colorup="green", colordn="red"),
    Volume(),
    Reversal(backcandels=2, forwardcandels=2, pivot_limit=100, show_pivots=True, scan_props=scan_props),
    # SMA(50),
    # SMA(200),
    RSI(),
    # MACD(),
]

chart = Chart(title=ticker, figsize=(20, 8))
chart.plot(prices, indicators)
```
