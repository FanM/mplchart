# Mplchart Typical Usage

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import OHLC, Candlesticks, Volume
from mplchart.indicators import SMA, RSI, MACD
```

## Fetching the prices

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices
```

## Creating the Chart

```python
style = "tableau"
max_bars = 250

indicators = [
    Candlesticks(),
    SMA(50),
    SMA(200),
    Volume(),
    RSI(),
    MACD(),
]

chart = Chart(title=ticker, max_bars=max_bars, style=style)
chart.plot(prices, indicators)
```
