```python
from mplchart.chart import Chart
from mplchart.helper import get_prices

from mplchart.primitives import Candlesticks, OHLC, Price, Volume
```

## Fetching the prices

```python
ticker = 'NVDA'
freq = 'daily'
prices = get_prices(ticker, freq=freq)
prices
```

## Candlestics plot with Candlesticks()

```python
max_bars = 250
indicators = [ Candlesticks() ]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Open High Low close bars with OHLC()

```python
max_bars = 250
indicators = [ OHLC() ]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Simple price plot with Price()

```python
max_bars = 250
indicators = [ Price() ]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Volume plot with Volume()

```python
max_bars = 250
indicators = [ Volume() ]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```

## Price and Volume combined

```python
max_bars = 250
indicators = [ Price(), Volume() ]

chart = Chart(title=ticker, max_bars=max_bars, figsize=(12, 9))
chart.plot(prices, indicators)
```
