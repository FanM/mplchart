# Mplchart Tests

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume
from mplchart.indicators import SMA, EMA, WMA, HMA, ROC, RSI, MACD
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices
```

```python
max_bars = 50

indicators = [
    Candlesticks(),
    EMA(20),
    HMA(20),
    Volume(),
    RSI(),
    MACD(),
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
max_bars = 50

indicators = [
    Candlesticks(),
    EMA(20),
    HMA(20),
    Volume(),
    RSI(),
    MACD(),
    ROC(1) @ HMA(20)
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
