# Talib Examples

```python
import pandas as pd
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, OHLC, Volume

from talib.abstract import Function
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices = prices.rename(columns=str.lower).rename_axis(index=str.lower)
prices.info()
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    Function("SMA", 50),
    Function("SMA", 200),
    Function("RSI"),
    Function("MACD"),
]

chart = Chart(title="Sample", max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    Function("BBANDS", 20),
    Function("PPO", 12, 26),
]

chart = Chart(title="Sample", max_bars=max_bars)
chart.plot(prices, indicators)
```
