# Fetch and save sample prices data (sample-prices.csv)

```python
import yfinance as yf

from pathlib import Path

ROOTDIR = Path.cwd().parent
```

```python
ticker = "AAPL"
prices = yf.download(ticker, period="max", auto_adjust=True, progress=False)
prices = prices.filter(["Open", "High", "Low", "Close", "Volume"])
prices = prices.rename(columns=str.lower).rename_axis(index=str.lower)
prices
```

```python
data = prices.to_csv(lineterminator="\n")
data
```

```python
outfile = ROOTDIR.joinpath("src/mplchart/samples/sample-prices.csv")
print(f"Updating {outfile.name} ...")
outfile.write_text(data)
```
