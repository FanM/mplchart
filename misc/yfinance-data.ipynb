# Candlesticks Chart

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks
```

```python
# intervals: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo

ticker, interval, period = "AAPL", "1m", "5d"

prices = yf.Ticker(ticker).history(interval=interval, period=period)
prices.info()
```

```python
max_bars = 250

indicators = [ Candlesticks(use_bars=True) ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
max_bars = 250

indicators = [ Candlesticks(colorup="b", colordn="r") ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
max_bars = 250

indicators = [ Candlesticks(colorup="b", colordn="r", use_bars=True) ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```
