# Selecting target axes with axes primitives

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume, NewAxes, SameAxes
from mplchart.indicators import SMA, EMA, WMA, HMA, ROC, RSI, MACD, DMI

```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Keeping two indicators in the same axes

max_bars = 250

indicators = [
    Candlesticks(),
    ROC(20),
    ROC(50) | SameAxes(),
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
# Plotting the SMA indicator in a separate axes

max_bars = 250

indicators = [
    Candlesticks(),
    SMA(20) | NewAxes(),
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
