```python


def get_settings(style=(), name=None):
    KEYS = ("lines.color", "text.color", "axes.edgecolor", "axes.facecolor",
            "figure.edgecolor", "figure.facecolor", "patch.edgecolor" )
    name = name or str(style)
    with plt.style.context(style):
        result = {
            k: plt.rcParams[k]
            for k in KEYS
            } | dict(name=name)
    return result

get_settings()
```

```python
style = { "lines.color": "pink" }
get_settings(style, "pink")
```

```python
records = [get_settings(style) for style in mstyle.available
           if not style.startswith("_")]


records = pd.DataFrame.from_records(records).set_index("name")
records
```
