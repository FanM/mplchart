# Testing chart bgcolor

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume

from mplchart.indicators import SMA, RSI, MACD
from mplchart.stylesheet import STYLESHEET
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    SMA(50),
    SMA(200),
    Volume(),
    RSI(),
    MACD(),
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
chart = Chart(title=ticker, max_bars=max_bars, bgcolor=None)
chart.plot(prices, indicators)
```

```python


chart = Chart(title=ticker, max_bars=max_bars, bgcolor="b")
chart.plot(prices, indicators)
```

```python
stylesheet = STYLESHEET | {"chart:bgcolor": "red"}
stylesheet
```

```python
chart = Chart(title=ticker, max_bars=max_bars, bgcolor=None, stylesheet=stylesheet)
print(chart.get_setting("chart", "bgcolor", default=None))
print(chart.get_stylemap().get_setting("chart", "bgcolor", default=None))
chart.plot(prices, indicators)
```
