# Color Scheme (experimental)

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume
from mplchart.indicators import SMA, EMA, WMA, HMA, ROC, RSI, MACD, DMI
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
color_scheme = dict(pdi="~green", ndi="~red", sma=["~r", "~g", "~b"])

max_bars = 250

indicators = [
    Candlesticks(),
    SMA(20),
    SMA(50),
    SMA(200),
    Volume(),
    RSI(),
    DMI(),
]

chart = Chart(title=ticker, max_bars=max_bars, color_scheme=color_scheme)

for name in "pdi", "ndi", "sma", "sma", "sma":
    color = chart.get_color(name, None)
    print(name, color)


chart.plot(prices, indicators)
```
