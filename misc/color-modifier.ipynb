# Color Modifier (experimental)

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume
from mplchart.indicators import SMA, EMA, WMA, HMA, ROC, RSI, MACD, DMI, BBANDS

from mplchart.modifiers import Color
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    SMA(20) | Color("orange"),
    MACD() | Color(macd="blue", macdsignal="red", macdhist="green")

]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
vars(SMA(20) | Color("pink"))
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    BBANDS() | Color("orange")

]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
