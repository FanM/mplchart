# Axes Modifiers (preliminary)

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume
from mplchart.indicators import SMA, EMA, WMA, HMA, ROC, RSI, MACD, DMI

from mplchart.modifiers import SameAxes, NewAxes
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    SMA(20) | NewAxes(),
    SMA(50),
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
vars(SMA(20) | NewAxes("below"))
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    ROC(20) | NewAxes(),
    ROC(50) | SameAxes(),
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
