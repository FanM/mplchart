# Matplotlib Cycler (playground)

```python
import numpy as np

import matplotlib.pyplot as plt



# See https://matplotlib.org/stable/users/explain/artists/color_cycle.html
```

```python

x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 4, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
yy.shape
```

```python
cmyk_cycler = plt.cycler(color=['c', 'm', 'y', 'k'])

style = {"axes.prop_cycle": cmyk_cycler}
# style = {}

with plt.style.context(style):
    ax = plt.axes()
    ax.set_title('Using CYMK color cycler')
    # ax.set_prop_cycle(cmyk_cycler)
    ax.plot(yy)
    plt.show()
```
