# LinePlot Primitive

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Price, Volume, LinePlot
from mplchart.indicators import SMA, EMA, RSI, ROC, ATRP, MACD, DMI
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    SMA(20) | LinePlot(style="dashed", color="red", alpha=0.5, width=3)
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
max_bars = 50

indicators = [
    Candlesticks(),
    SMA(20) | LinePlot(style="marker", color="red", alpha=0.5)
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```
