```python
def main_settings():
    for key in ("axes.edgecolor", "axes.facecolor",  "lines.color", ):
        value = plt.rcParams[key]
        print(key, value)

main_settings()
```

```python
pink_lines = { "lines.color": "pink" }
with plt.style.context(pink_lines):
    main_settings()
```

```python
with plt.style.context("classic"):
    main_settings()
```

```python
with plt.style.context({}):
    main_settings()
```

```python
for style in plt.style.available:
    print(f"*** {style} ***")
    with plt.style.context("classic"):
        main_settings()
    print("\n")
```
