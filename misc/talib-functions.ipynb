# Using ta-lib abstract functions with mplchart

```python
import pandas as pd
import yfinance as yf

from talib import get_functions
from talib.abstract import Function

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks
```

```python
functions = [Function(name).info for name in get_functions()]
functions = pd.DataFrame(functions).set_index("name")
functions
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices = prices.rename(columns=str.lower).rename_axis(index=str.lower)
prices
```

```python
max_bars = 50
indicators = [
    Candlesticks(),
    Function("SMA", 50),
    Function("SMA", 200),
    Function("KAMA"),
    Function("BBANDS"),
    Function("RSI"),
    Function("MACD"),
]

chart = Chart(title="Sample", max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
sma50 = Function("SMA", 50)
sma50
```

```python
sma50.function_flags
```

```python
from mplchart.extalib import TalibWrapper

wrapper = TalibWrapper(sma50)
wrapper
```
